```python
# Import necessary libraries
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt

# Step 1: Load the Dataset
file_path = 'audio_1.csv'  # Replace with your dataset path
audio_data = pd.read_csv(file_path)

# Step 2: Clean and Prepare the Data
# Select the relevant features and include track_id for reference
features = ['track_id', 'danceability', 'energy', 'acousticness', 'instrumentalness', 'valence', 'tempo']
audio_data_clean = audio_data[features].dropna()

# Extract track_id for later use
track_ids = audio_data_clean['track_id']

# Drop track_id from the features used in clustering
features_for_clustering = audio_data_clean.drop(columns=['track_id'])

# Step 3: Standardize the Data
scaler = StandardScaler()
audio_data_scaled = scaler.fit_transform(features_for_clustering)

# Step 4: Apply KMeans Clustering
n_clusters = 3  # Adjust this as needed
kmeans = KMeans(n_clusters=n_clusters, random_state=42)
clusters = kmeans.fit_predict(audio_data_scaled)

# Add cluster labels back to the original dataset with track_id
audio_data_clean['Cluster'] = clusters

# Step 5: Analyze Cluster Characteristics
# Calculate mean only for numeric columns
cluster_centroids = audio_data_clean.groupby('Cluster').mean(numeric_only=True)

print("Cluster Characteristics:")
print(cluster_centroids)

# Map human-readable cluster labels
cluster_labels = {
    0: 'Energetic Dance Tracks',
    1: 'Relaxing Acoustic Tunes',
    2: 'Upbeat Happy Songs'
}
audio_data_clean['Cluster_Label'] = audio_data_clean['Cluster'].map(cluster_labels)

# Step 6: Visualize Clusters in 2D
pca = PCA(n_components=2)
pca_result = pca.fit_transform(audio_data_scaled)

plt.figure(figsize=(10, 6))
plt.scatter(pca_result[:, 0], pca_result[:, 1], c=clusters, cmap='viridis', s=50)
plt.title('Clusters Visualized (PCA)')
plt.xlabel('PCA Dimension 1')
plt.ylabel('PCA Dimension 2')
plt.colorbar(label='Cluster')
plt.show()

# Step 7: Generate Playlists with track_id
playlists = {}
for cluster_label in audio_data_clean['Cluster_Label'].unique():
    playlist_name = f'Playlist_{cluster_label.replace(" ", "_")}'
    playlist = audio_data_clean[audio_data_clean['Cluster_Label'] == cluster_label]
    playlists[playlist_name] = playlist
    # Save each playlist with track_id included
    playlist.to_csv(f'{playlist_name}.csv', index=False)
    print(f"Saved {playlist_name} with {len(playlist)} songs.")

# Step 8: Recommend Similar Songs (Track ID Included)
#from scipy.spatial.distance import cdist

#def recommend_similar_with_ids(song_index, num_recommendations=5):
    # Find similar songs within the same cluster
    selected_song = audio_data_scaled[song_index].reshape(1, -1)
    cluster_id = clusters[song_index]
    cluster_data = audio_data_scaled[clusters == cluster_id]
    distances = cdist(selected_song, cluster_data, metric='euclidean')
    closest_indices = distances.argsort()[0][:num_recommendations]
    similar_songs = audio_data_clean.iloc[closest_indices][['track_id', 'Cluster_Label']]
    return similar_songs

## Example: Recommend 5 similar songs to the first song
recommendations = recommend_similar_with_ids(song_index=0, num_recommendations=5)
print("Recommended Songs (with Track IDs):")
print(recommendations)
```
